```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/home/infinity/Desktop/QA-Project/Data/Cleaned/BankChurners_cleaned.csv")

df.head()
df.info()
df.describe()
```

```python
# بيطلع عدد الناس اللي مشيت واللي لسه موجوده والسطر اللي بعده بيجبها ب نسبه مئويه
df['Attrition_Flag'].value_counts()
df['Attrition_Flag'].value_counts(normalize=True)
```

```python
sns.countplot(x='Attrition_Flag', data=df)
plt.show()
```

```python
#Numerical Features Distribution
# بختار شويه  اعمده تكون ارقام وبرسمهم
# this phase is Univariate Analysis  remember it 
num_cols = [
    'Customer_Age','Credit_Limit','Total_Trans_Amt',
    'Total_Trans_Ct','Avg_Utilization_Ratio'
]

df[num_cols].hist(figsize=(12,8), bins=20)
plt.show()
```

```python
#Categorical Features Analysis
# this phase is  Bivariate analysis
# هنا انا بعمل فيتشر مقابل التارجت 
sns.countplot(x='Gender', hue='Attrition_Flag', data=df)
plt.show()
```

```python
sns.countplot(x='Education_Level', hue='Attrition_Flag', data=df)
plt.show()
```

```python
#Relationship
sns.boxplot(x='Attrition_Flag', y='Total_Trans_Amt', data=df)
plt.show()

sns.boxplot(x='Attrition_Flag', y='Total_Trans_Ct', data=df)
plt.show()
```

```python
#Behavior Analysis
sns.boxplot(x='Attrition_Flag', y='Avg_Utilization_Ratio', data=df)
plt.show()

sns.boxplot(x='Attrition_Flag', y='Credit_Limit', data=df)
plt.show()
```

```python
#Correlation Matrix
#Multivariate Analysis
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.show()
```
