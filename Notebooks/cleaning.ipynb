```python
def split_income_range(value):
    if value == "Unknown":
        return pd.Series([np.nan, np.nan])
    
    if value == "Less than $40K":
        return pd.Series([0, 40000])
    
    if value == "$120K +":
        return pd.Series([120000, np.nan])   # end is open
    
    # For ranges like "$60K - $80K"
    parts = value.replace("$", "").replace("K", "000").split(" - ")
    start = int(parts[0])
    end = int(parts[1])
    return pd.Series([start, end])
```

```python
df[["Income_Start", "Income_End"]] = df["Income_Category"].apply(split_income_range)
```

```python
df["Income_Start"] = df["Income_Start"].fillna(df["Income_Start"].median())
df.loc[df["Income_Category"] == "$120K +", "Income_End"] = 150000

```

```python
df = df.drop(columns=["Income_Category"])
```
